```python
# check the same articles
def same_element(train_data, test_data):
    col_list = train_data.columns.tolist()
    for i in col_list:
        set1 = set(train_data[i])
        set2 = set(test_data[i])
        print(i, 'same: ', (set1 & set2), i, 'same number: ', len(set1 & set2))
```

```python
mp_title = pd.DataFrame(most_popular_json_results, columns=['title'])
shared_title = pd.DataFrame(shared_results, columns=['title'])
same_element(mp_title, shared_title)
```
