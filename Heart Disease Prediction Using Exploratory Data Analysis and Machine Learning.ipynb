## Phase 2 Predictive Model:

```python
# load the LogisticRegression() function from sklearn's 'linear_model' sub-library
from sklearn.linear_model import LogisticRegression
```

```python
data
```

```python
# change the value of Diagnosis to 0 or 1. 0 is no heart disease, 1 is heart disease
data["Diagnosis"] = np.where(data["Diagnosis"] == 0, 0, 1)
# change the value of PainType to 0 or 1. 0 is asymptotic, 1 is anginal pain
data["PainType"] = np.where(data["PainType"] == 4, 0, 1)
# change the value of RestECG to 0 or 1. 0 is normal, 1 1 is with abnormality
data["RestECG"] = np.where(data["RestECG"] == 0, 0, 1)
# change the value of STPeakSlope to 0 or 1. 0 is upsloping, 1 is flat or downsloping
data["STPeakSlope"] = np.where(data["STPeakSlope"] == 1, 0, 1)
data
```

```python
# a randomly selected 80% of the data from the dataset is used as training data
train = data.sample(frac = 0.8)
train
```

```python
# the rest of the data is used as test data.
test = data.drop(train['Diagnosis'].index)
test
```

```python
len(train) + len(test)
```

```python
# define a vector containing the names of the attributes to use
predictors = ['Age', 'Gender', 'PainType', 'BloodPres', 'Chol', 'FBSugar', 'RestECG', 'MaxHeartRate', 'ExAngina', 'STPeakSlope']

# create a subset of the training data using the selected explanatory variables
X_train = train[predictors].values

# create a subset of the testing data using the selected explanatory variables
X_test = test[predictors].values

# isolate the response indicator for the training data
y_train = train['Diagnosis'].values

# sanity check on training data
X_train[:5]
```

```python
# sanity check on response indicator
y_train[:5]
```

```python
# use the LogisticRegression() method for this model
model = LogisticRegression()

# fit the model
model.fit(X_train, y_train)
```

```python
# calculate the accuracy of the model relative to the training data set
model.score(X_train, y_train)
```

```python
# examine the model coefficients for the explanatory variables
print(predictors)
model.coef_
```

In summary, we can see that:

- The effect of age is less. The older people are, the less likely people are to have a heart attack.
- Men are more likely to have heart disease than women. And the influence of gender is very large, it's the second largest of these factors.
- Angina is the most influential factor. asymptotic angina is associated with a higher likelihood of heart disease..
- The effect of resting blood pressure is less. People with higher blood pressure are also more likely to develop heart disease.
- Cholesterol is the most uninfluential factor. The higher the cholesterol, the increased likelihood of heart disease.
- The effect of fasting blood sugar is less. The higher the fasting blood sugar, the increased the likelihood of heart diseas.
- People with abnormal resting ECG have an increased likelihood of having heart disease.
- The effect of max heart rate achieved is less. The higher the maximum heart rate, the higher the likelihood of having heart disease.
- The effect of exercise induced angina is big. People who have ever had angina induced by exercise will be more likely to have heart disease than other people.
- The effect of Peak exercise ST segment is big. Abnormalities in the Peak exercise ST segment can lead to a higher likelihood of heart disease.

```python
# generate predictions for the test data using our new model
y_predict = model.predict(X_test)
y_predict[:10]
```

There are some conclusion that contradict reality, such as the effect of age on the likelihood of heart disease. For further analysis, we can perform regression analysis based on the existing findings to improve our feature engineering and create better predictive models.
