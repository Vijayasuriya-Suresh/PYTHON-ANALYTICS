# Exploratory Data Analysis -- AQI

An air quality index (AQI) is used by government agencies to communicate to the public how polluted the air currently is or how polluted it is forecast to become. AQI information is obtained by averaging readings from an air quality sensor, which can increase due to vehicle traffic, forest fires, or anything that can increase air pollution. Pollutants tested include ozone, nitrogen dioxide, sulphur dioxide, among others.

Public health risks increase as the AQI rises, especially affecting children, the elderly, and individuals with respiratory or cardiovascular issues. During these times, governmental bodies generally encourage people to reduce physical activity outdoors, or even avoid going out altogether. The use of face masks such as cloth masks may also be recommended.

The United States Environmental Protection Agency (EPA) has developed an Air Quality Index that is used to report air quality. This AQI is divided into six categories indicating increasing levels of health concern. An AQI value over 300 represents hazardous air quality and below 50 the air quality is good.

### Import air quality data from web API

```python
# Import the required libraries
import pandas as pd
import requests
import json
from pandas.io.json import json_normalize

# Returns an array of the Air Quality
air_quality = requests.get('https://aqs.epa.gov/data/api/dailyData/byCounty?email=mwang3@mail.yu.edu&key=indigoswift45&param=88101&bdate=20210101&edate=20211231&state=36&county=061')
# API status code
print(air_quality.status_code)

import requests
from pprint import pprint

def airquality(query):
    res=requests.get('https://aqs.epa.gov/data/api/dailyData/byCounty?email=mwang3@mail.yu.edu&key=indigoswift45&param=88101&bdate=20210101&edate=20211231&state=36&county=061');
    return res.json();

# Retrieved data of the Air Quality
air_quality = airquality(air_quality)
air_quality = air_quality['Data']
air_quality
```

## Correlation

The correlation coefficient is an indicator of the degree of linear correlation between two variables (the square of the correlation coefficient is called the coefficient of determination). Correlation analysis refers to the analysis of variables to determine the degree of correlation between two variable factors, and then to influence the other factor by guiding one of them.
There are three main methods commonly used to measure correlation between variables.

- Pearson correlation coefficient: i.e. Pearson correlation coefficient, which is used to measure the correlation coefficient between two continuous random variables.
- Spearman correlation coefficient: i.e., Spearman correlation coefficient, which is a rank correlation coefficient. It is solved according to the ranking position of the original data and is also defined as the Pearson correlation coefficient between rank variables.
- Kendall correlation coefficient: i.e., Kendall correlation coefficient, which is also a rank correlation coefficient and is often calculated for ordered categorical variables.

Because our dataset has all continuous numeric variables except for the date variable. So we will use Pearson correlation coefficient correlation coefficient.

For the correlation analysis, there are two values to note. One is the correlation coefficient r and the other is the significance level p. The relationship between the two is such that the r value can only be referred to when p is less than 0.05 (or 0.01), and not just the r value. Assuming p=0.02 and r=0.8, the two sets of data are considered to have a highly linear relationship. Assuming p=0.5 and r=0.8, it is considered that the two sets of data cannot be compared in correlation (the correlation performance at this point may be coincidental).

In general, we consider that:
- an absolute value of r of 0.3, when no correlation exists;
- the absolute value of r is greater than 0.3 and less than 0.5, at which point a low correlation exists;
- the absolute value of r is greater than 0.5 and less than 0.8, at which point there is a significant correlation;
- the absolute value of r is greater than 0.8, at which time there exists a high correlation.
